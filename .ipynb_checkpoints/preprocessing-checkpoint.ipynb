```python
def clean_disconti(array):
    clean_arr = np.empty((len(array)))
    for i in range(len(array)):
        if (i>2) & (i<len(array)-2):
            if min(abs(array[i]-array[i+1]), abs(array[i]-array[i-1])) > 0.7:
#                 idx = list(abs(array[i-2:i+3]-np.mean(array[i-2:i+3]))).index(min(abs(array[i-2:i+3]-np.mean(array[i-2:i+3]))))
#                 clean_arr[i] = array[i-2:i+3][idx]
                clean_arr[i] = (array[i-1]+array[i+1])/2
            else:
                clean_arr[i] = array[i]
#             elif max(abs(array[i-2]-array[i-1]), abs(array[i+1]-array[i+2])) < max(abs(array[i]-array[i+1]), abs(array[i]-array[i-1])):
#                 array[i] = (array[i+1]+array[i-1])/2
        else:
            clean_arr[i] = array[i]
    return clean_arr
```

```python
def find_criteria(data, percentage):
    statastic = plt.hist(data, bins = 1000)[0]/len(data)
    bin_num = plt.hist(data, bins = 1000)[1]
    plt.clf()
    sum_stat = 0
    for i in range(len(statastic)):
        sum_stat += statastic[i] 
        if sum_stat > percentage:
            criteria = bin_num[i]
            break
    return criteria
# plt.xlim((0,0.2))
# plt.show()
```

```python
time_interval = []
start = 1
for i in range(len(jump_candidate_time)):
    if (i>start) & (i+1<len(jump_candidate_data)-1):
        sub_time = []
        for j in range(i,len(jump_candidate_data)-2):
            if (jump_candidate_data[j]-jump_candidate_data[j-1])*(jump_candidate_data[j+1]-jump_candidate_data[j])>0:
                if (markers_on_first_der[j] not in sub_time) :
                    sub_time.append(markers_on_first_der[j])
                if (markers_on_first_der[j-1] not in sub_time) :
                    sub_time.append(markers_on_first_der[j-1])
                if (markers_on_first_der[j+1] not in sub_time) :
                    sub_time.append(markers_on_first_der[j+1])
#                 print(i, sub_time)
            else:
                start = j+1
                break
        if len(sub_time)>0:
            time_interval.append(sub_time)
         
```

```python
time_interval
```

```python
time_dur = [i[-1]-i[0] for i in time_interval ]
fixing_criteria = find_criteria(time_dur, 0.99)
fixing_criteria
```

```python
time_point_raw = [[i[0], i[-1]] for i in time_interval]
```

```python
time_point = []
for index in range(len(time_point_raw)):
    if index < len(time_point_raw)-2:
        if time_point_raw[index+1][0]-time_point_raw[index][-1] < fixing_criteria:
            time_point.append([time_point_raw[index][0], time_point_raw[index+1][-1]])
        else:
            time_point.append(time_point_raw[index])
            
```

```python
def linear(X, y):
    m = (y[-1]-y[0])/(X[1]-X[0])
    cor = m*(np.linspace(X[0], X[1], (X[1]-X[0])+1)-X[0])+y[0]
    return cor 
```

```python
fix_jump = np.copy(modify_temp_watch)
for time_index in time_point:
    
    start = time_index[0]
    end = time_index[1]
    if end-start <85:  
        fix_jump[start-5:end+5] = [fix_jump[start-6] for i in range(len(fix_jump[start-5:end+5]))]
    else:
        fix_jump[start-5:end+5] = linear([start-5,end+4], fix_jump[start-5:end+5])
```
